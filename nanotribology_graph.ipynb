```python
import numpy as np
import mypylib as mpl
import importlib
importlib.reload(mpl)
output_path = 'output/fig3_1.png'

def x_0(x, k):
    return (2 * np.pi / k) * np.cos(2 * np.pi * x) + x

# 初期設定
x = np.linspace(0, 5, 400)
k_init = 50

x0 = x_0(x, k_init)

# 図の作成
data1 = [x0, x, 'black']
data = [data1]
xrange = [-2,6]
yrange = [min(x), max(x)]
mpl.line_graph(data, '$x_0$', '$x$', '', xrange, yrange, output_path)
```

```python
import numpy as np
import mypylib as mpl
import matplotlib.pyplot as plt

output_path = 'output/fig3_4.png'

def f_int(x):
    return np.sin(x)

# 初期設定
x = np.linspace(-4, 10, 1000)
y = f_int(x)

# 図の作成
# グラフの描画
plt.figure(figsize=(5, 5))
plt.plot(x, y, color = 'black')
# y=0 の点線を追加
plt.axhline(0, color='gray', linestyle='dashed', linewidth=1)
plt.tick_params(direction='in', which='both')  # 'both' でx軸とy軸両方の目盛りを内向きに
# ラベルと凡例
plt.xlabel("$x_0$",fontsize=15)
plt.ylabel("$F$",fontsize=15)
plt.xlim(-4, 10)
plt.savefig(output_path,bbox_inches='tight')
plt.show()
```
