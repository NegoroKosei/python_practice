```python
class Polygon:
    WIDTH = 0
    HEIGHT = 0
    @classmethod
    def set_wh(cls, width, height):
        cls.WIDTH = width
        cls.HEIGHT = height
    @classmethod
    def input_wh(cls):
        cls.WIDTH = float(input ('幅を入力してください : '))
        cls.HEIGHT = float(input ('高さを入力してください : '))
    @classmethod
    def calc_area(cls):
        pass
    @classmethod
    def print_are(cls):
        pass

class Rectangle(Polygon):
    @classmethod
    def calc_area(cls):
        return cls.WIDTH * cls.HEIGHT
    @classmethod
    def print_area(cls):
        print(f"四角形の面積は {cls.calc_area()}")

class Triangle(Polygon):
    @classmethod
    def calc_area(cls):
        return cls.WIDTH * cls.HEIGHT * 0.5
    @classmethod
    def print_area(cls):
        print(f"直角三角形の面積は {cls.calc_area()}")
```

```python
# コンソールからの入力
# Rectangle.input_wh()
# 直接入力
Rectangle.set_wh(0.1,4)
Rectangle.print_area()
Triangle.print_area()

```

```python
# インスタンスを利用した計算方法
class Polygon:
    # コンストラクタ、共通の要素は名前のみ
    def __init__(self, name):
        self.name = name
    # 長さの入力
    def set_l(self):
        pass
    # コンソールからの入力
    def input_l(self):
        pass
    # 面積計算
    def calc_area(self):
        pass
    # 面積計算+出力
    def print_area(self):
        pass

# Plygonを継承
class Rectangle(Polygon):
    def __init__(self, width = 0, height = 0):
        self.width = width
        self.height = height
        super().__init__("rectangle")
    def set_l(self, width, height):
        self.width = width
        self.height = height
    def input_l(self):
        self.width = float(input ('幅を入力してください : '))
        self.height = float(input ('高さを入力してください : '))
    def calc_area(self):
        return self.width * self.height
    def print_area(self):
        print(f"{self.name}の面積は {self.calc_area()}")

class Triangle(Polygon):
    def __init__(self, width, height):
        self.width = width
        self.height = height
        super().__init__("triangle")
    def set_l(self, width, height):
        self.width = width
        self.height = height
    def input_l(self):
        self.width = float(input ('幅を入力してください : '))
        self.height = float(input ('高さを入力してください : '))
    def calc_area(self):
        return self.width * self.height * 0.5
    def print_area(self):
        print(f"{self.name}の面積は {self.calc_area()}")

class Circle(Polygon):
    PI = 3.14159265
    def __init__(self, radius = 0):
        self.radius = radius
        super().__init__("circle")
    def set_l(self, radius):
        self.radius = radius
    def input_l(self):
        self.radius = float(input ('半径を入力してください : '))
    def calc_area(self):
        return self.radius * self.radius * Polygon.PI
    def print_area(self):
        print(f"{self.name}の面積は {self.calc_area()}")

rectangle01 = Rectangle()
rectangle01.set_l(2.0, 4.0)
rectangle01.print_area()

triangle01 = Triangle(2.0, 5.0)
triangle01.print_area()

circle01 = Circle()
circle01.input_l()
circle01.print_area()

# 共通要素を抜き出してできるfor文を回せる
poly01 = [rectangle01, triangle01, circle01]
for p in poly01:
    p.print_area()
```

```python
# インターフェイスを利用した計算方法
from abc import ABC, abstractmethod
PI = 3.14159265
class PolygonInterface(ABC):
    # 長さの入力
    @abstractmethod
    def set_l(self):
        pass
    # コンソールからの入力
    @abstractmethod
    def input_l(self):
        pass
    # 面積計算
    @abstractmethod
    def calc_area(self):
        pass
    # 面積計算+出力
    @abstractmethod
    def print_area(self):
        pass

# PolygonInterfaceを引き継ぎ
class Rectangle(PolygonInterface):
    def __init__(self, width = 0, height = 0):
        self.width = width
        self.height = height
        self.name = "rectangle"
    def set_l(self, width, height):
        self.width = width
        self.height = height
    def input_l(self):
        self.width = float(input ('幅を入力してください : '))
        self.height = float(input ('高さを入力してください : '))
    def calc_area(self):
        return self.width * self.height
    def print_area(self):
        print(f"{self.name}の面積は {self.calc_area()}")

class Triangle(PolygonInterface):
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.name = "triangle"
    def set_l(self, width, height):
        self.width = width
        self.height = height
    def input_l(self):
        self.width = float(input ('幅を入力してください : '))
        self.height = float(input ('高さを入力してください : '))
    def calc_area(self):
        return self.width * self.height * 0.5
    def print_area(self):
        print(f"{self.name}の面積は {self.calc_area()}")

class Circle(PolygonInterface):
    def __init__(self, radius = 0):
        self.radius = radius
        self.name = "circle"
    def set_l(self, radius):
        self.radius = radius
    def input_l(self):
        self.radius = float(input ('半径を入力してください : '))
    def calc_area(self):
        return self.radius * self.radius * PI
    def print_area(self):
        print(f"{self.name}の面積は {self.calc_area()}")

rectangle01 = Rectangle()
rectangle01.set_l(2.0, 4.0)
rectangle01.print_area()

triangle01 = Triangle(2.0, 5.0)
triangle01.print_area()

circle01 = Circle()
circle01.input_l()
circle01.print_area()

# 共通要素を抜き出してできるfor文を回せる
poly01 = [rectangle01, triangle01, circle01]
for p in poly01:
    p.print_area()
```
