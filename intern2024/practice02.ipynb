```python
# 生徒管理クラス
# クラスの名前には動詞を使わない
class Student:
    # クラス変数
    SCHOOL_NAME = "電通大"

    # クラスメソッドを示すでコーター(@関数に性質を与える,毎回前につける)
    @ classmethod
    def get_area(cls,j,k):
        cls.SCHOOL_NAME = "親クラス関数よばれた大学"
        return j*k
    @ classmethod
    def get_area2(cls,j,k):
        cls.SCHOOL_NAME = "電通大"
        return j*k

    # コンストラクタ(__init__)
    def __init__(self, id, name):
        print("Student初期化メソッド実行:")

        # 変数の前に__をつけることでprivateに、ただし厳密なprivateではない
        # "__id"はクラス外で"_Student__id"に名前マグリング(そういう仕組み)によって変換されているだけ
        self.__id = id
        self.name = name
    def print_info(self):
        print("Studentのprint_info:",self.SCHOOL_NAME,self.__id,self.name)

    def get_id(self):
        return self.__id

# プログラムが直接実行されたときのみ使用
if __name__=="__main__":
    a = Student(100, "太郎")
    a.print_info()

    # idには直接アクセスできない
    print("直接print:",Student.SCHOOL_NAME,a.get_id())
    # 名前マグリングの仕組みを知っていればアクセスできる
    print("直接print:",a._Student__id)

    # クラスのインスタンスからクラスメソッド,クラス変数を呼べる
    print(a.get_area(2, 3))
    print(a.SCHOOL_NAME)

    # クラス名からも呼べる
    print(Student.get_area2(2, 3))
    print(Student.SCHOOL_NAME)
    # インスタンスメソッドもオブジェクトを渡すことで実行できる
    # この時、インスタンスメソッドをクラスメソッドのように用いる
    Student.print_info(a)
```

```python
# 継承 ... class A (B):
# private変数へはアクセスできない
# 親クラスのインスタンス変数にはアクセスできない

# 記述すると、クラス関数、クラス変数はオーバーライドされる
# オーバーライドすると、別のものとして扱われる
# 何も書かないとクラス関数、クラス変数は継承される
# 親のクラス変数、クラス関数を変えると子のクラス変数、関数も変わる(逆は不成立)
# アクセスするときは(クラス名). を用いる

class IStudent(Student):
    # SCHOOL_NAME = "電通大2"
    def __init__ (self, id, name, nationality):
        print("IStudent初期化メソッド実行:")
        self.nationality = nationality

        # スーパークラスの呼び出し。※ ()を忘れない
        # これでオーバーライド前の関数も呼び出せる
        super().__init__(id,name)
    def print_info(self):
            print("IStudentのprint_info:",self.SCHOOL_NAME,self.get_id(),self.name, self.nationality)

a =IStudent(102,"Syo","FN")
print( "直接プリント:", a.name, a.nationality)
a.print_info()

# 関数を直接指定することでオーバーライド前の親クラスの関数も呼び出せる
Student.print_info(a)

print()
# 親クラスのクラス変数をいじる→オーバーライドがなければともに変化
print(Student.get_area(2, 3))
print("親クラスのクラス変数:",Student.SCHOOL_NAME)
print("子クラスのクラス変数:",IStudent.SCHOOL_NAME)

# 子クラスのクラス変数をいじる→子クラスのみ変化
# print(IStudent.get_area(2, 3))
# print("親クラスのクラス変数:",Student.SCHOOL_NAME)
# print("子クラスのクラス変数:",IStudent.SCHOOL_NAME)
```
